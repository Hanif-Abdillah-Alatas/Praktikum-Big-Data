```python
Task 1: Insert a dataset of courses where each course has multiple student enrollments, then display the data.
```

```python
from pymongo import MongoClient, InsertOne

# Establish client connection
client = MongoClient('mongodb://localhost:27017/')
db = client['university_db']
courses_collection = db['courses']
courses_collection.delete_many({})

# Bulk insert of 20 engineering courses
operations = [
    # Computer Engineering
    InsertOne({'course': 'Intro to Programming', 'enrollments': 40, 'department': 'Computer Engineering'}),
    InsertOne({'course': 'Data Structures', 'enrollments': 32, 'department': 'Computer Engineering'}),
    InsertOne({'course': 'Computer Networks', 'enrollments': 28, 'department': 'Computer Engineering'}),
    InsertOne({'course': 'Operating Systems', 'enrollments': 26, 'department': 'Computer Engineering'}),

    # Electrical Engineering
    InsertOne({'course': 'Circuit Analysis', 'enrollments': 35, 'department': 'Electrical Engineering'}),
    InsertOne({'course': 'Electromagnetics', 'enrollments': 22, 'department': 'Electrical Engineering'}),
    InsertOne({'course': 'Control Systems', 'enrollments': 30, 'department': 'Electrical Engineering'}),
    InsertOne({'course': 'Digital Signal Processing', 'enrollments': 27, 'department': 'Electrical Engineering'}),

    # Mechanical Engineering
    InsertOne({'course': 'Thermodynamics', 'enrollments': 29, 'department': 'Mechanical Engineering'}),
    InsertOne({'course': 'Fluid Mechanics', 'enrollments': 31, 'department': 'Mechanical Engineering'}),
    InsertOne({'course': 'Dynamics of Machinery', 'enrollments': 18, 'department': 'Mechanical Engineering'}),
    InsertOne({'course': 'Heat Transfer', 'enrollments': 34, 'department': 'Mechanical Engineering'}),

    # Civil Engineering
    InsertOne({'course': 'Structural Analysis', 'enrollments': 24, 'department': 'Civil Engineering'}),
    InsertOne({'course': 'Construction Materials', 'enrollments': 20, 'department': 'Civil Engineering'}),
    InsertOne({'course': 'Transportation Engineering', 'enrollments': 33, 'department': 'Civil Engineering'}),
    InsertOne({'course': 'Geotechnical Engineering', 'enrollments': 19, 'department': 'Civil Engineering'}),

    # Chemical Engineering
    InsertOne({'course': 'Organic Process Design', 'enrollments': 36, 'department': 'Chemical Engineering'}),
    InsertOne({'course': 'Chemical Reaction Engineering', 'enrollments': 28, 'department': 'Chemical Engineering'}),
    InsertOne({'course': 'Process Control', 'enrollments': 26, 'department': 'Chemical Engineering'}),
    InsertOne({'course': 'Industrial Chemistry', 'enrollments': 21, 'department': 'Chemical Engineering'}),
]

# Execute bulk write
courses_collection.bulk_write(operations)
print('Courses inserted successfully.\n')

# Display the data
for course in courses_collection.find():
    print(f"Course: {course['course']}, Enrollments: {course['enrollments']}, Department: {course['department']}")
```

```python
Exercise 2: Filter courses with enrollments over 20 students.
```

```python
# Query for courses with enrollments greater than 20
for course in courses_collection.find({'enrollments': {'$gt': 20}}):
    print(course)
```

```python
Task 2: Retrieve and display courses only from the 'Computer Engineering' or 'Civil Engineering' departments.
```

```python
# Query courses in Computer or Civil Engineering departments
for course in courses_collection.find({'department': {'$in': ['Computer Engineering', 'Civil Engineering']}}):
    print(course)
```

```python
Exercise 3: Use aggregation to find the average enrollment per department.
```

```python
# Average enrollment per department using aggregation
pipeline = [
    {'$group': {'_id': '$department', 'average_enrollment': {'$avg': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
Task 3: Create an aggregation pipeline that finds the maximum enrollment for each department.
```

```python
# Maximum enrollment per department
pipeline = [
    {'$group': {'_id': '$department', 'max_enrollment': {'$max': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
Exercise 4: Use $project to rename and only show fields: course_name, department_name, and enrollments.
```

```python
# Projection to rename fields
pipeline = [
    {'$project': {'course_name': '$course', 'department_name': '$department', 'enrollments': 1}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
Task 4: Use $addFields to create a new field enrollment_category where enrollments > 25 are 'high' and others 'low'.
```

```python
# Adding enrollment category field based on enrollments
pipeline = [
    {'$addFields': {'enrollment_category': {'$cond': {'if': {'$gt': ['$enrollments', 25]}, 'then': 'high', 'else': 'low'}}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
Homework 1: Perform an aggregation to get a count of courses per department.
```

```python
# Count courses per department
pipeline = [
    {'$group': {'_id': '$department', 'total_courses': {'$sum': 1}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
Homework 2: Use $match and $group together to filter and get only courses with enrollments over 25 in 'Computer Engineering'.
```

```python
# Filter courses with enrollments over 25 in computer engineering department
pipeline = [
    {'$match': {'department': 'Computer Engineering', 'enrollments': {'$gt': 25}}},
    {'$group': {'_id': '$department', 'total_courses': {'$sum': 1}, 'avg_enrollments': {'$avg': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(f"Department: {result['_id']}, Total Courses: {result['total_courses']}, Average Enrollments: {result['avg_enrollments']:.2f}")
```

```python
Homework 3: Apply $lookup to join courses collection with students collection based on student enrollments.
```

```python
# Join courses collection with students collection
pipeline = [
    {'$lookup': {'from': 'students', 'localField': 'course', 'foreignField': 'course', 'as': 'enrolled_students'}},

]
for result in courses_collection.aggregate(pipeline):
    print(result)
```
