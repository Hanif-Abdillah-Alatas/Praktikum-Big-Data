Praktik PySpark Lanjutan

```python
from pyspark.sql import SparkSession

# Memulai Spark Session
spark = SparkSession.builder.appName('BigDataPractice').getOrCreate()

# Membuat DataFrame
data = [("Hanif", 18), ("Dana", 20), ("Andra", 31), ("Nayla", 25)]
columns = ["Nama", "Umur"]
df = spark.createDataFrame(data, columns) # df: DataFrame

# Menampilkan
df.show()

# Filter Data
filtered_df = df.filter(df.Umur > 20) # Menampilkan Data Dengan Umur > 20
filtered_df.show()

# Rata-Rata (AVG) Umur
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Umur")).show()

# Mengurutkan Data Sesuai Umur (Turun)
df.sorted = df.orderBy("Umur", ascending=False)
df.sorted.show()
```

```python
from pyspark.sql import SparkSession

# Memulai Spark Session
spark = SparkSession.builder.appName('BigDataPractice').getOrCreate()

# Membuat DataFrame
data = [("Hanif", 18, "Mahasiswa", "Bermain", "Laki-laki"),
        ("Dana", 25, "Hacker", "Coding", "Laki-laki"),
        ("Andra", 36, "Pelukis", "Menggambar", "Laki-laki"),
        ("Nayla", 27, "Akuntan", "Jalan-jalan", "Perempuan")]
columns = ["Nama", "Umur", "Pekerjaan", "Hobi", "Kelamin"]
df = spark.createDataFrame(data, columns) # df: DataFrame

# Menampilkan
print("Data Masyarakat")
df.show()

# Filter Data
filtered_df = df.filter(df.Kelamin == "Laki-laki") # Menampilkan Data Dengan Kelamin Laki-Laki
print("Data Masyarakat Yang Laki-Laki")
filtered_df.show()

# Rata-Rata (AVG) Umur
from pyspark.sql.functions import avg
print("Rata-Rata Umur Masyarakat")
df.groupBy().agg(avg("Umur")).show()

# Mengurutkan Data Sesuai Umur (Naik)
print("Data Masyarakat Sesuai Umur")
df.sorted = df.orderBy("Umur", ascending=True)
df.sorted.show()
```

```python
import pandas as pd
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('BigDataPractice').getOrCreate()

# DataFrame Spark
data_spark = [("Hanif", 18), ("Dana", 20), ("Andra", 31), ("Nayla", 25)]
columns = ["Nama", "Umur"]
df_spark = spark.createDataFrame(data_spark, columns)

# DataFrame Pandas
data_pandas = {"Nama": ["Bagus", "Gayuh", "Salwa", "Nadine"],
               "Umur": [26, 30, 17, 29]}
df_pandas = pd.DataFrame(data_pandas)

# Konversi Spark ke Pandas
df_pandas_from_spark = df_spark.toPandas()

# Konversi Pandas ke Spark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

print("DataFrame Konversi Spark ke Pandas:")
print(df_pandas_from_spark)

print("\nDataFrame Konversi Pandas ke Spark:")
df_spark_from_pandas.show()

# Menggabungkan Data
df_joined = pd.concat([df_pandas_from_spark, df_pandas])
print("Data Gabungan:")
print(df_joined)

# Rata-rata Umur
rata_rata = df_joined["Umur"].mean()
print(f"\nRata-Rata Umur: {rata_rata}")
```

```python
import pandas as pd
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('BigDataPractice').getOrCreate()

# DataFrame Spark
data_spark = [("Hanif", 18), ("Dana", 20), ("Andra", 31), ("Nayla", 25)]
columns = ["Nama", "Umur"]
df_spark = spark.createDataFrame(data_spark, columns)

# DataFrame Pandas
data_pandas = {"Nama": ["Bagus", "Gayuh", "Salwa", "Nadine"],
               "Umur": [26, 30, 17, 29]}
df_pandas = pd.DataFrame(data_pandas)

# Konversi Spark ke Pandas
df_pandas_from_spark = df_spark.toPandas()

# Konversi Pandas ke Spark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

print("DataFrame Konversi Spark ke Pandas:")
print(df_pandas_from_spark)

print("\nDataFrame Konversi Pandas ke Spark:")
df_spark_from_pandas.show()

# Menggabungkan Data
df_joined = pd.concat([df_pandas_from_spark, df_pandas])
print("Data Gabungan:")
print(df_joined)

# Nilai Umur Maksimal (paling besar)
nilai_max = df_joined["Umur"].max()
print(f"\nNilai Maksimum Umur: {nilai_max}")
```
